# Inner Join

```python
import_df = pd.read_csv("/kaggle/input/india-trade-data/2018-2010_import.csv",header=0)
export_df = pd.read_csv("/kaggle/input/india-trade-data/2018-2010_export.csv",header=0)
```

```python
import_df.head()
```

```python
export_df.head()
```

```python
import_df.HSCode.describe()
```

```python
e_Query1 = """
           SELECT * 
           FROM export_df 
           WHERE HSCode<30
           """
```

```python
df_e1 = sqldf(e_Query1)
```

```python
df_e1.head()
```

```python
df_e1.tail()
```

```python
df_e1.shape
```

```python
i_Query1 = """
           SELECT * 
           FROM import_df
           WHERE HSCode<30
            """
```

```python
df_i1 = sqldf(i_Query1)
```

```python
df_i1.head()
```

```python
df_i1.tail()
```

```python
df_i1.HSCode.values
```

```python
df_i1.year.unique()
```

```python
join_Query = """
             SELECT im.HSCode, im.value as import_Values, e.value as export_Values, im.year
             FROM 
                 import_df im
             INNER JOIN 
                 export_df e
                     ON im.HSCode = e.HSCode
             GROUP BY im.HSCode
             ORDER BY im.year
             LIMIT 10;
             
             """
```

```python
df_join=sqldf(join_Query)
```
